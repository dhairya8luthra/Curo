```python
X = df.drop(columns=["PremiumPrice"]).values
Y=df["PremiumPrice"].values
```

```python
X_train,Y_train,X_test,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)

print(X_train.shape)
print(Y_train.shape)
```

```python
from sklearn.preprocessing import StandardScaler

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

print(X_train.shape)
print(Y_train.shape)

from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
predictions1 = logreg.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print("Confusion Matrix: \n", confusion_matrix(Y_test, predictions1))
print('\n')
print(classification_report(Y_test, predictions1))
```

```python
from sklearn.metrics import accuracy_score

logreg_acc = accuracy_score(Y_test, predictions1)
print("Accuracy of the Logistic Regression Model is: ", logreg_acc)
```

```python
from sklearn.neighbors import KNeighborsClassifier

error_rate = []

for i in range(1,42):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, Y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != Y_test))
```

```python
plt.figure(figsize=(12,6))
plt.plot(range(1,42), error_rate, color='purple', linestyle="--",
         marker='o', markersize=10, markerfacecolor='b')
plt.title('Error_Rate vs K-value')
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=9)
knn.fit(X_train, Y_train)
predictions2 = knn.predict(X_test)
```

```python
print(confusion_matrix(Y_test, predictions2))
print("\n")
print(classification_report(Y_test, predictions2))
```

```python
from sklearn.metrics import accuracy_score
knn_model_acc = accuracy_score(Y_test, predictions2)
print("Accuracy of K Neighbors Classifier Model is: ", knn_model_acc)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=300)
rfc.fit(X_train, Y_train)
predictions4 = rfc.predict(X_test)
```

```python
print("Confusion Matrix: \n", confusion_matrix(Y_test, predictions4))
print("\n")
print(classification_report(Y_test, predictions4))
```

```python
from sklearn.svm import SVC

svc_model = SVC(kernel="rbf")
svc_model.fit(X_train, Y_train)
predictions5 = svc_model.predict(X_test)
```

```python
print("Confusion Matrix: \n", confusion_matrix(Y_test, predictions5))
print("\n")
print(classification_report(Y_test, predictions5))
```

```python
svm_acc = accuracy_score(Y_test, predictions5)
print("Accuracy of SVM model is: ", svm_acc)
```

```python
import lightgbm as lgb
clf = lgb.LGBMClassifier()
clf.fit(X_train, Y_train)
```

```python
y_pred=clf.predict(X_test)
accuracy=accuracy_score(y_pred, Y_test)
print('LightGBM Model accuracy score: {0:0.4f}'.format(accuracy_score(Y_test, y_pred)))
```

```python
print("Logistic Regression Score: ",logreg_acc)
print("KNN Score: ",knn_model_acc)
print("RandomForest Score: ",rfc_acc)
print("SVM Score: ",svm_acc)
print("LightGBM Score: ", accuracy)
```

```python
plt.figure(figsize=(12,6))
model_acc = [logreg_acc, knn_model_acc, rfc_acc, svm_acc,accuracy]
model_name = ['LogisticRegression', 'KNN', 'RandomForests', 'SVM',"LightGBM"]
sns.barplot(x= model_acc, y=model_name, palette='magma')
```

```python
from skl2onnx import convert_sklearn
from skl2onnx.common.data_types import FloatTensorType

# Define the initial type - note we have 10 features (float32)
initial_type_scaler = [('input', FloatTensorType([None, 10]))]
scaler_onnx = convert_sklearn(ss, initial_types=initial_type_scaler)

# Save the scaler model to a file.
with open("scaler.onnx", "wb") as f:
    f.write(scaler_onnx.SerializeToString())
    
print("Preprocessing ONNX model exported as 'scaler.onnx'.")
```

```python
# Convert the trained LightGBM model into an ONNX model.
# First, register a shape calculator and converter for LGBMClassifier.
from skl2onnx import update_registered_converter
from skl2onnx.common.shape_calculator import calculate_linear_classifier_output_shapes
import onnxmltools
from skl2onnx.common.data_types import FloatTensorType

# Define the initial type: we have 10 features, as float values.
initial_type_model = [('input', FloatTensorType([1, 10]))]


options = {id(clf): {'zipmap': False}}
# Convert the LightGBM model using onnxmltools.
onnx_model = onnxmltools.convert_lightgbm(clf, initial_types=initial_type_model,zipmap=False)

# Save the converted ONNX model
with open("model.onnx", "wb") as f:
    f.write(onnx_model.SerializeToString())

print("Prediction ONNX model exported as 'model.onnx'.")
```
